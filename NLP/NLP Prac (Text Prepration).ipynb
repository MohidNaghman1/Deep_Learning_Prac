## Remove Html tags and URLs in the text

```python
import re

text = "Welcome to <b>my website</b>!"
cleaned = re.sub(r'<[^>]+>', '', text)
print(cleaned)
```

```python
def remove_html(text):
    return re.sub(r'<[^>]+>', '', text)

def remove_urls(text):
    return re.sub(r'http[s]?://\S+|www\.\S+', '', text)

def clean_text(text):
    if isinstance(text, str):
        text = remove_html(text)
        text = remove_urls(text)
    return text
```

```python
df = df.applymap(clean_text)
```

## Remove Punctuation

```python
import string
exclude = string.punctuation
```

```python
def remove_punctuation(text):
    # Create a translation table that maps each punctuation character to None
    translator = str.maketrans('', '', exclude)
    return text.translate(translator)
```

```python
df = df.applymap(remove_punctuation)
```

## Spelling Correction

```python
from textblob import TextBlob

def correct_spelling(text):
    # Create a TextBlob object
    blob = TextBlob(text)
    return str(blob.correct())
```

```python
df = df.map(correct_spelling)
```

## Remove Stopwords

```python
# pip install nltk
```

```python
import nltk
from nltk.corpus import stopwords

# Download the list of stopwords
nltk.download('stopwords')

# Get English stopwords
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    # Split text into words
    words = text.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)
```

```python
df = df.map(remove_stopwords)
```

```python
import spacy

# Load the English language model
nlp = spacy.load('en_core_web_sm')
```

```python
def spacy_tokenization(text):
    # Process the text through spaCy pipeline
    doc = nlp(text)
    # Extract tokens from the processed document (doc)
    return [token.text for token in doc]
```

```python
df['token'] = df['text'].apply(spacy_tokenization)
```

## Stemming

```python
from nltk.stem import PorterStemmer
```

```python
stm = PorterStemmer()

def apply_stemming(tokens):

    return [stm.stem(token) for token in tokens]
```

```python
df['token'] = df['token'].apply(apply_stemming)
```
